```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

qr = QuantumRegister(1)
cr = ClassicalRegister(1)

# The quantum circuit takes in the quantum register and classical register
qc = QuantumCircuit(qr, cr)
qc.initialize([3/5, 4/5], 0)
qc.measure(qr, cr)
qc.draw(output="mpl")
```

```python
from qiskit_aer import AerSimulator
from qiskit import *

backend = AerSimulator()
job = backend.run(qc, shots=100, memory=True)

from qiskit.tools.visualization import plot_histogram

result = job.result()
counts = result.get_counts()

plot_histogram(counts)
```

```python
import matplotlib

mem = result.get_memory()

def fun(a,c):
    d = [i for i,x in enumerate(a) if x == c]
    e = [(x+1)/(i+1) for x, i in enumerate(d)]
    matplotlib.pyplot.plot(d, e)

fun(mem, '1')
fun(mem, '0')
```
